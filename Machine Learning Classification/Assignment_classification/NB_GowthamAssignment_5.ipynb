```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
indep=dataset[["age", "bp", "al", "su", "bgr", "bu", "sc", "sod", "pot", "hrmo", "pcv", "wc", "rc"]]
dep=dataset["classification"]
```

```python
indep.shape
```

```python
dep
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 0.10, random_state = 0)
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

X_train_enc = encoder.fit_transform(X_train)
X_test_enc = encoder.transform(X_test)
```

```python
from sklearn.naive_bayes import CategoricalNB
classifier =CategoricalNB(alpha=1.0)
classifier.fit(X_train_enc, y_train)
y_pred = classifier.predict(X_test_enc)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, y_pred)              
print(clf_report)
print(cm)
```
